```python
def a(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return a(n-1) + a(n-2)
for i in range(0, 7):
    print(a(i), end="")
```

```python
def fact(n):
    n_fact = 1
    for i in range(1, n+1):
        n_fact = n_fact * i
    return n_fact

n = int(input())
fact(n)
print(fact(n))
b  = fact(n) ** 2
print(b)
```
